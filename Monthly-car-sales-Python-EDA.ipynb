```python
#importing necessary libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='dark')
import warnings 
warnings.filterwarnings('ignore')
```

```python
#loading the data
data = pd.read_csv('Monthly-car-sales.csv', parse_dates=True, index_col=0, header=0, infer_datetime_format=True)
data.head()
```

```python
#Obtaining a descriptive statistics of the dataset
round(data.describe(), 1)
```

```python
#checking for null valuess
data.isnull().sum()
```

```python
data.plot()
plt.grid()
```

```python
#A histogram helps to understand the distribution of the data points
plt.hist(data, density=True, color='r')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid()
```

```python
#plotting a boxplot to better understand the distribution
plt.boxplot(data, vert=False)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Boxplot')
plt.grid()
```

```python
from scipy.stats import skew
skew = skew(data)

print('The skewness of MSFT close prices is:', skew)

if skew > 0.00:
    print('The dataset is positively skewed')
else:
    print('The dataset is negatively skewed')
```

```python
data.to_csv('Monthly_car_sales.csv')
```
